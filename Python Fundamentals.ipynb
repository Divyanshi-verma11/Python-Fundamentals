```python
class Fraction():
    # Parameterized constructor
    def __init__(self, num, dem):
        self.num = num
        self.dem = dem
    # Return how the object exactly look
    def __str__(self):
        return "{}/{}".format(self.num, self.dem)
        
    def Hcf(self, num, dem):
        hcf = 1
        if num == dem:
            hcf = num
        else:
            for i in range(1, min(num, dem) + 1):
                if num % i == 0 and dem % i == 0:
                    hcf *= i
                    num = num / i
                    dem = dem / i
        return hcf

    def check(self, num, dem):
        for i in range(2, min(num, dem)):
                if num % i == 0 and dem % i == 0:
                    num = num / i
                    dem = dem / i
        return int(num), int(dem)
        
    def __add__(self, a):
        self.num, self.dem = self.check(self.num, self.dem)
        a.num, a.dem = self.check(a.num, a.dem)
        num = self.num * a.dem  + a.num * self.dem
        dem = self.dem * a.dem
        x = self.Hcf(num, dem)
        return "{}/{}".format(int(num/x) , int(dem/x))

    def  __sub__(self, a):
        self.num, self.dem = self.check(self.num, self.dem)
        a.num, a.dem = self.check(a.num, a.dem)
        num = self.num * a.dem  - a.num * self.dem
        dem = self.dem * a.dem
        x = self.Hcf(num, dem)
        if num == 0:
            return 0
        else:
            return "{}/{}".format(int(num/x) , int(dem/x))

    def __mul__(self, a):
        self.num, self.dem = self.check(self.num, self.dem)
        a.num, a.dem = self.check(a.num, a.dem)
        num = a.num * self.num
        dem = a.dem * self.dem
        x = self.Hcf(num, dem)
        return "{}/{}".format(int(num/x) , int(dem/x))

    def __truediv__(self, a):
         self.num, self.dem = self.check(self.num, self.dem)
         a.num, a.dem = self.check(a.num, a.dem)
         num = a.dem * self.num
         dem = a.num * self.dem
         x = self.Hcf(num, dem)
         return "{}/{}".format(int(num/x) , int(dem/x))
    def convert_to_decimal(self):
        return self.num / self.dem
f1 = Fraction(30, 60) 
f2 = Fraction(20, 40)
print(f1 + f2)
print(f1 - f2)
print(f1 * f2)
print(f1 / f2)
f1.convert_to_decimal()
```

# Write OOP classes to handle the following scenarios

* A user can create and view 2D coodinates
* A user can find the distance between 2 coodinates
* A user can find the distance of a coodinates from origin
* A user can check if a point lies on a given line
* A user can find the distance betwenn a given 2D point and given line

```python
class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return "<{}, {}>".format(self.x, self.y)

    def euclidean_distance(self, a):
        d = ((a.x - self.x)**2 + (a.y - self.y)**2)**0.5
        return round(d, 2)

    def distance_from_origin(self):
        # d = ((self.x)**2 + (self.y)**2)**0.5
        d = self.euclidean_distance(Point(0, 0))
        return round(d, 2)

class Line():
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def __str__(self):
        if self.c >= 0:
            return "{}x + {}y + {} = 0".format(self.a, self.b,self.c)
        else:
            return "{}x + {}y - {} = 0".format(self.a, self.b,-self.c)

    def check(line,  point):
        if line.a * point.x + line.b * point.y + line.c == 0:
            return "Exist on the line"
        else:
            return "Doesnot exist on the line"

    def distance_of_point_from_line(line, point):
        d = abs((line.a*point.x + line.b * point.y + line.c))/(line.a**2 + line.b**2)**0.5
        return round(d, 2)
```

# Inheritance in summary
* A class can inherit from another class.
* Inheritance improves code reuse.
* Constructor, attributes, methods get inherited to the child class.
* The parent has no access to the child class.
* Private properties of parent are not accessible directly in child class.
* Child class can override the attributes or methods. This is called method overriding.
* super() is an inbuilt function which is used to invoke the parent class methods and constructor.

```python
class MyException(Exception):
    def __init__(self, message):
        print(message)
        
def sanity_check(data_type):
    def check(func):
        def wrapper(*args):
            if type(*args) == data_type:
                func(*args)
            else:
                raise MyException("Invalid datatype")
        return wrapper
    return check

@sanity_check(int)
def square(num):
    print(num ** 2)
    
@sanity_check(str)
def names(name):
    print("Namaste", name)
try:
    square(3)
    names("Divyanshi")
except MyException as e:
    pass
```

# Abstract notion of order of growth

what do we want:-
* We want to evaluate the algorithm
* We want to eavluate scalability
* we want to evaluate the terms of input size

Order a Growth
* want to evaluate program’s efficiency when input is very big
* want to express the growth of program’s run time as input size grows
* want to put an upper bound on growth — as tight as possible
* do not need to be precise: “order of” not “exact” growth
* we will look at largest factors in run time (which section of the program will take the longest to run?)
* thus, generally we want tight upper bound on growth, as function of size of input, in worst case

Types of Orders of Growth
* Constant (searching index in array)   [O(1)]
* Linear (searching random number)   [O(log n)]
* Quadratic (nested loops)                    [O(n)]
* Logarithmic (Binary search)              [O(n log n)]
* nlogn (Sorting algorithm Merge sort)[O(n ^ 2)]
* Exponential (2 ** n)                           [O(2 ^ n)]

* Law if Addition(Independent program)
* Law of Multiplication(nested loops)

```python
class NLPApp:
    def __init__(self):
        self.__database = {}
        self.__second_menu()
    def __first_menu(self):
        users = input("""
        1. Resgiter
        2. Login
        3. Exit
        """)
        if users == "1":
            self.__register()
        elif users == "2":
            self.__login()
        else:
            print("You Exit...!!!!")
            return
            
    def __second_menu(self):
        u = input("""
        Hiii! How you like to proceed....???
        1. NER
        2. Language Detection
        3. Sentiment Analysis
        4. Exit
        """)
        try: 
            if u == "1":
                self.__ner()
            elif u == "2":
                self.__Language_detection()
            elif u == "3":
                self.__sentiment_analysis()
            else:
                print("You Exit...!!!!")
                return
        except Exception as e:
            print(e)
    def __register(self):
        name = input("Enter the name.... ")
        email = input("Enter the email..... ")
        password = input("Enter the password..... ")
        if email in self.__database:
            print("Email is already exist!!! ")
            self.__register()
        else:
            self.__database[email] = [name, password]
            print("Successfully Resgiter.....")
            print()
            self.__first_menu()
            
    def __login(self):
        email = input("Enter the email..... ")
        password = input("Enter the password..... ")
        if email in self.__database:
            if self.__database[email][1] == password:
                print("Login Successful")
                self.__second_menu()
            else:
                print("Password is wrong")
                self.__login()
        else:
            n = input("Want to re enter the Email......")
            if n == "yes":
                self.__login()
            else:
                print("Register first...... ")
                self.__register()
    def __ner(self):
        para = input("Enter a paragraph...... ")
        print()
        lang = input("What would you like to search..... ")
        client = nlpcloud.Client("gpt-oss-120b", "6485f76081cc7a8df0ae36ab947697a2da9a473a", gpu=True)
        response = client.entities(para, searched_entity=lang)
        print()
        print(response)
        print()
        return self.__second_menu()
        
    def __Language_detection(self):
        para = input("Enter a paragraph...... ")
        print()
        client = nlpcloud.Client("python-langdetect", "6485f76081cc7a8df0ae36ab947697a2da9a473a", gpu=False)
        res = client.langdetection(para)
        print()
        print(res)
        print()
        return self.__second_menu()

    def __sentiment_analysis(self):
        para = input("Enter a paragraph...... ")
        print()
        client = nlpcloud.Client("gpt-oss-120b", "6485f76081cc7a8df0ae36ab947697a2da9a473a", gpu=True)
        response = client.sentiment(para, target="NLP cloud")
        L = []
        for i in response['scored_labels']:
            L.append(i['score'])
            
        index = sorted(list(enumerate(L)), key=lambda x: x[1], reverse=True)[0][0]
        
        print(response['scored_labels'][index]['label'])
        return self.__second_menu()
```

# Order of growth

```python
def power(num):
    if num < 0:
        return 0
    elif num == 1:
        print(1)
        return 1
    else:
        prev = power(num//2)
        current = prev * 2
        print(current)
        return current
power(10)
print("Complete")
 # time complexity O(Log(n))
```

```python
def mod(m, n):
    if n < 0:
        return -1
    div = m // n
    return m - div * n
mod(5, 4)
 # time complexity O(1)
```

```python
def powset(l, index = 0, subset = [[]]):
    if index == len(l):
        return subset
        
    sub = []
    
    for i in subset:
        copy_set = i.copy()
        copy_set.append(l[index])
        sub.append(copy_set)
    subset.extend(sub)
    return powset(l, index + 1, subset)
    
l = [1, 2, 3]
powset(l)
 # time complexity O(2 ^ n)
```

```python
def fibo(n, d):
    if n in d:
        return d[n]
    else:
        d[n] = fibo(n - 1, d) + fibo(n - 2, d)
        return d[n]
d = {0 : 1, 1 : 1}
fibo(1000, d)
 # time complexity O(n)
```
