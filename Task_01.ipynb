### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
leg = int(input("Enter a Leg.... ")) # input leg numbers
head = int(input("Enter a head.... ")) # input head numbers
if head > leg or head == 0 or leg % 2 != 0: # wrong input
  print("No solution")

else:
  d = (leg - 2 * head)/2
  c = head - d
  print(f"number of dogs : {d}, number of chickens : {c}")
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
first = int(input("enter first number.... ")) # enter first number
second = int(input("Enter second number.... ")) # enter second number
n  = int(input("Enter Nth position... ")) # enter nth position
d = second - first # difference value
print(first + (n - 1) * d) # display value of nth position
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
t_h = int(input("Enter height of tank..... ")) # height of the tank
t_l = int(input("Enter length of tank..... ")) # length of the tank
t_b = int(input("Enter breadth of tank..... ")) # breadth of the tank
g_h = int(input("Enter height of glass..... ")) # height of the glass
g_r = int(input("Enter radius of glass..... ")) # radius of the glass
v = t_h * t_l * t_b # volumn of the tank
g = 3.14 * g_r * g_r * g_h # volumne of the glass
print("Number of glass obtained.... ", v//g) # number of glass obtained
````
